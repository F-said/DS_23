**Parametric Method**
We assume some sort of form to our distribution, i.e. normal distribution

* Parametric does not always assume normal!
* Most methods that we have at our disposal assume normal, so we default to this distribution
* We usually apply these methods on large datasets.
* We only apply these methods to continuous variables (numbers)
* This assumption of distribution lends to its statistical power

**Non-Parametric Methods**
We cannot assume any form to our distribution (distribution-free)

* We usually have a small sample size 
* Since we cannot assume distribution, we have less statistical power
* Variables could be discrete (labels) or continuous (numbers)

```python
import random

total_heads = 0
flips = 100

for flip in range(flips):
    random_choice = random.randint(0,1)
    if random_choice == 0:
        total_heads += 1

print(f"The probability of getting heads is {total_heads/flips}")
```

## LAW OF LARGE NUMBERS


```python
import random
import matplotlib.pyplot as plt

x_axis = []
probs = []

for trials in range(1, 10_000, 10):
    total_heads = 0
    flips = trials
    x_axis.append(flips)

    for flip in range(flips):
        random_choice = random.randint(0,1)
        if random_choice == 0:
            total_heads += 1
    
    probs.append(total_heads/flips * 100)

plt.xkcd()
plt.tight_layout()

plt.plot(x_axis, probs)
plt.axhline(y = 50, color = 'r', linestyle = '-')

plt.title("# Trials vs probability of heads")
plt.xlabel("# Of Trials")
plt.ylabel("Probability of Heads")
plt.ylim([0, 100])

plt.savefig("LAWLARGE_coins.jpg")
```

```python
import random
import matplotlib.pyplot as plt

x_axis = []
probs = []

for trials in range(1, 10_000, 10):
    total = 0
    rolls = trials
    x_axis.append(rolls)

    for roll in range(rolls):
        total += random.randint(1,6)
    
    probs.append(total/rolls)

plt.xkcd()
plt.tight_layout()

plt.plot(x_axis, probs)
plt.axhline(y = 3.5, color = 'r', linestyle = '-')

plt.title("# Trials vs Avg Dice Roll")
plt.xlabel("# Of Trials")
plt.ylabel("Average of Dice Rolls")

plt.savefig("LAWLARGE_dice.jpg")
```
