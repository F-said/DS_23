## Probability

There are two major camps of probability:
* Probability is a linguistic & mathematical tool to explain the chances of something happening.
> Ex: I have a coin. It has two sides. If I flip it, I have a 50% chance of getting heads because there are only 1 out of 2 outcomes.
> ½ = 0.50 = 50%
* Probability is the ratio of the number of desired outcomes out of all possible outcomes.
> I have a coin. I flip it 100 times and take note of the number of times heads popped up.
> Heads showed up 51 times. Therefore my probability is 51%

We prefer to use both definitions, since as we conduct more trials in our experiments experimental probability approaches our theoretical

## Coin Flips


Keep in mind, this is the **law of large numbers**: 

Let's say we run an experiment where we first flip a coin 10 times, and then 110 times, and then 210 times. How will our experimental probability look as we conduct more trials?

```python
import random

total_heads = 0
flips = 100

for flip in range(flips):
    random_choice = random.randint(0,1)
    if random_choice == 0:
        total_heads += 1

print(f"The probability of getting heads is {total_heads/flips}")
```

## LAW OF LARGE NUMBERS 

As we conduct more trials in our experiment, our experimental probability approaches our theoretical value!

```python
import random
import matplotlib.pyplot as plt

x_axis = []
probs = []

for trials in range(1, 10_000, 10):
    total_heads = 0
    flips = trials
    x_axis.append(flips)

    for flip in range(flips):
        random_choice = random.randint(0,1)
        if random_choice == 0:
            total_heads += 1
    
    probs.append(total_heads/flips * 100)

plt.xkcd()
plt.tight_layout()

plt.plot(x_axis, probs)
plt.axhline(y = 50, color = 'r', linestyle = '-')

plt.title("# Trials vs probability of heads")
plt.xlabel("# Of Trials")
plt.ylabel("Probability of Heads")
plt.ylim([0, 100])
```

```python
import random
import matplotlib.pyplot as plt

x_axis = []
probs = []

for trials in range(1, 10_000, 10):
    total = 0
    rolls = trials
    x_axis.append(rolls)

    for roll in range(rolls):
        total += random.randint(1,6)
    
    probs.append(total/rolls)

plt.xkcd()
plt.tight_layout()

plt.plot(x_axis, probs)
plt.axhline(y = 3.5, color = 'r', linestyle = '-')

plt.title("# Trials vs Avg Dice Roll")
plt.xlabel("# Of Trials")
plt.ylabel("Average of Dice Rolls")
```

## Basic Probability

There are a couple of rules that we should know:

* `P(A) = Outcomes of A / All Outcomes`
* `P(Ac) = 1 - P(Ac)`
* `P(A & B) = P(A) * P(B)`
* `P(A U B) = P(A) + P(B)`


## Dependent Probability

* `P(A U B) = P(A) + P(B) - P(A & B)`
* `P(A|B) = P(B|A)P(A) / P(B|A)P(A) + P(B|Ac)P(Ac)`

## Bayes Theorem

Bayes theorem is our approach to calculating conditional probability, with prior observationa.

Example:

You've tested yourself for COVID and recieved a positive test. What is the probability that you have COVID given that the accuracy rate is 96% & COVID is prevelant in 35% of US Adults?

`P(A|B) = (0.96 * 0.35) / (0.96 * 0.35) + (0.04 * 0.65)`
`P(A|B) = 0.92`

You can now use this as a prior probability if you were to test positive a second time!

`P(A|B) = (0.96 * 0.35) / (0.96 * 0.92) + (0.04 * 0.08)`
`P(A|B) = 0.99`

After your second positive COVID test, you have a 99% chance of having covid!

## Skewness & Kurtosis

**Skewness**: Is there a tail going to the left or to the right? That is, are their highly positive outliers or highly negative outliers? Positive skew indicates outliers to right, negative skew indicates outliers to left. unitless

**Excess Kurtosis**: peakedness of distribution compared to normal. Highly positive kurtosis is high peakedness, very negative is low peakeedness, and 0 is normal. Also unitless. Positive kurtosis usually also indicates high outliers, whereas negative kurtosis indicates no outliers. 0 indicates normality.

## Normal Distributions

Most things in life are getting to in this distribution when considering quantitative frequency. For most of our ML methods, we want to get as close as possible to this!

Normal distributions indicate that data is evenly distributed around the mean.

Excess Kurtosis = 0, Skew = 0

```python
import scipy.stats
import matplotlib.pyplot as plt


num_samples = 10000

plt.rcdefaults()
sample = scipy.stats.norm.rvs(size=num_samples)
plt.hist(sample, bins=100)
plt.show()
```

```python
scipy.stats.kstest(sample, scipy.stats.norm.cdf, args=(sample.mean(), sample.std()))

# notice the p-value of this ks-test for normality!
```

## Hypothesis Testing

Let’s review the hypothesis tests we’ve interacted with so far:

* One-Sample T-test: compare mean of 1 dataset to theoretical value. Assume normality. 
* Two-Sample T-Test: compare means of 2 independent datasets. Assume normality.
* Paired-Sample T-test: compare means of 2 dependent datasets. Assume normality.
* Kolmogorov-Smirnov-Test: compare 2 datasets to check if are from same distribution. No normality assumed.

The p-value that these tests calculate is the probability that there is a difference between our two samples due to RANDOM CHANCE

If p < 0.05 → We reject the null hypothesis, as the probability there is a difference between the two samples is miniscule.

If p > 0.05 → We DO NOT reject the null hypothesis, as there is a good chance there is no difference between samples.

## Applications

* When should we use it? 
 * T-tests are for comparing means
 * KS-tests are for comparing distributions 
* What is our null hypothesis?
* Do we assume a distribution?
* What does our output indicate?
  * If p < 0.05, we deny the null hypothesis. What is the null hypothesis in this test?


```python
import random 
from scipy import stats
import scipy

# 1 = heads
# 0 = tails
unfair_coin = [1, 1, 1, 1, 1, 0, 1, 1, 0, 1]
fair_coin = [random.randint(0, 1) for _ in range(0, len(unfair_coin))]

print(sum(unfair_coin)/len(unfair_coin))
print(sum(fair_coin)/len(fair_coin))

print("t-test", stats.ttest_ind(unfair_coin, fair_coin))
```

## Central Limit Theorem

The central limit theorem states if you…
* have more than 30 random samples from a population
* and calculate the average of this dataset to get your sample mean
* and you repetitively keep doing this
* your collection of sample means WILL FORM A NORMAL DISTRIBUTION AROUND THE THEORITCAL MEAN 


This means that ANY metric that you calculate (as long as it is unbiased and > 30) from a sample, most likely falls within 2-3 standard deviations of your TRUE UNIVERSAL METRIC.

## Standard Error

How can we tell how far off our sample statistic (for example, sample mean) is away from our population mean? 

We use a metric called STANDARD ERROR.

This is the estimated *standard deviation of your sample mean from the population mean.*

We express this as 

`SE = (standard deviation)/sqrt(n)`

```python
population = scipy.stats.norm.rvs(size=num_samples)

sample = np.random.choice(population, size=30)

sample_mean = sample.mean()

plt.hist(population, bins=100)
plt.axvline(x = sample_mean, color = 'r', linestyle = '-')
plt.show()

print("The standard error of this sample is", sample.std()/(len(sample)**(1/2)))
```

## Confidence Intervals

Confidence intervals show us a range of possible values around the sample mean. Z is the level of confidence that we want from our experiment. Consider, if our cutoff for significance is 0.05, what is our confidence level?

`0.95`.

Confidence intervals calculate a range of values:

CI_0 = sample_mean - Z*(SE)  
CI_1 = sample_mean + Z*(SE)


```python
SE = sample.std()/(len(sample)**(1/2))

CI_0 = sample_mean - ((0.95)*SE)
CI_1 = sample_mean + ((0.95)*SE)

print(CI_0, CI_1)

plt.hist(population, bins=100)

plt.axvline(x = sample_mean, color = 'r', linestyle = '-')
plt.fill_betweenx([0, 350], CI_0, CI_1, color='g', alpha=0.5)

plt.show()
```

The green translucent bar represents the range of possible values (with 95% confidence) of the sample mean. Notice that the population mean falls within this confidence interval!

Once again, this is the central limit theorem revealing itself to us!

## Confidence Intervals as Indicators

When analyzing groups of data, and looking for significant differences in means, it's a good idea to calculate confidence intervals of your two groups!

If these confidence intervals overlap, you might just be looking at different means due to pure chance.

However, if confidence intervals do NOT overlap, you just might be looking at a significant difference between groups!

Keep in mind this is just the first start to exploring causality. We cannot definitely state that there is a causal relationship just yet!

## Z-Test

The Z-Test is a hypothesis test to check if the difference of two means is significant.  
    Null hypothesis → No statistical difference between means  
    Alt Hypothesis → There is a statistical difference between means  

Assumptions:
 * We have > 30 samples
 * Our dataset is normally distributed
 * Population standard deviation is known

Just like with t-tests, we have 3 different types of z-tests:
 * One-Sample
 * Two-Sample
 * Paired-Sample

Application: Same as t-test, to check the probability that a difference in means is due to random chance.

https://www.statsmodels.org/dev/generated/statsmodels.stats.weightstats.ztest.html 

While our dataset is not normally distributed (yet), we can take some liberties & state that with a large enough dataset, we can still apply these hypothesis tests.

```python
import scipy.stats as stats
from statsmodels.stats.weightstats import ztest as ztest
import pandas as pd
import numpy as np

clean_df = pd.read_csv("data/cleaned_data.csv")

# remove outliers from groups
# even after z-score standardization, we still do not see good normality (as evidenced by the our ks-test)
# this might hint that z-scores is not sufficient for extreme outliers
clean_df = clean_df[(np.abs(stats.zscore(clean_df["salary_standardized"])) < 3)]

# split into python & not python
df_python = clean_df[clean_df["python"] == 1]
df_nopython = clean_df[clean_df["python"] == 0]

# check normality of population
stats.kstest(clean_df["salary_standardized"], stats.norm.cdf, args=(clean_df["salary_standardized"].mean(), clean_df["salary_standardized"].std()))
```

## ANOVA

The Analysis of Variance (ANOVA) test is a hypothesis test to check if the difference between the means of 3 or more groups is statistically significant. (think of multiple t-tests for all possible pairs of groups)

    Null hypothesis → No statistical difference between means (all means are the same!)
    Alt Hypothesis → There is a statistical difference between means

Assumptions:
* Groups are independent 
* Our dataset is normally distributed
* Groups have similar variance (homogeneity of variances)

We have 2 different types of ANOVA tests:
* One-way ANOVA (one independent factor)
* Two-Way ANOVA (two independent factors)

Application: Check for co-variance between groups, which might reveal that some independent variables are actually DEPENDENT!

## Levene's Test

A levene-test is a hypothesis test to check for the homogeneity of variance between two groups

    Null hypothesis → Variance in two groups are equal (homogenous)
    Alt Hypothesis → Variance in two groups are NOT equal

Assumptions:
* We have > 30 samples
* Our dataset is normally distributed. However, this test is robust against non-normal data! 
* Datasets are independent

Application: Check if variance is statistically different between groups. Remember, we reject the null hypothesis if p < 0.05. Therefore, if p < 0.05, we state that there is NO HOMOGENEITY OF VARIANCE! We should not be doing simple ANOVA!

https://biotoolbox.binghamton.edu/Biostatistics/Biostatistics%20Student%20Tutorials/Levene%20Test/Levene.pdf

```python
stats.levene(df_python["salary_standardized"], df_sql["salary_standardized"], df_tab["salary_standardized"])
```

## Kruskal–Wallis ANOVA

The Kruskal-Wallis analysis of variance is a nonparametric analysis of variance designed for:
* Any distribution of data
* Small datasets
* Homogeneity of variance is not preserved

Null hypothesis → There is NO significant difference in means
Alt Hypothesis → There IS a  significant difference in means

Application: If homogeneity of variance is not preserved, but you still want to check if you have significant difference of means between groups, you can utilize the Kruskal-Wallis ANOVA test.
