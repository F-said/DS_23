## Cleaning

We managed to create our dataframe that only include movies in the `movies` DataFrame object

As we might have noticed, there are missing values in our dataframe.

Missing values could potentially bias our results, or block certain methods from running.

We will talk about more formal methods of data cleaning throughout this phase, but for now, let us simply delete any row that contains a null value.

We could potentially ‚Äúfilter‚Äù our rows using our filters.

Filtering should be reserved for DataFrame subsetting. We will use built-in functions.

Conceptually however, we also want to filter out data that only has more than 30 votes (Central Limit Theorem)

This is to avoid the possibility that only one loyal fan rated a movie 10/10 when in reality most people will find this film unpalatable.

How can we achieve this?


````python
movies.dropna(inplace=True)

movies_valid = movies[movies["vote"] > 30]
movies_valid
````

## Builtins

Now that have our dataframe prepared & cleaned. How can I answer the following question?
Is the count of movies increasing or decreasing?

We will use our built pandas methods, namely `value_counts(‚Äúcolumn‚Äù)`

This function works by counting how many unique instances it views in the specified column, and creates a new dataframe detailing this.

You‚Äôll notice that our years is not sorted by default, so we will call another builtin function called `sort_index`which sorts our dataframe according to its index.

Just observing this dataframe is not sufficient to visualize our differences, so we will actually be using a builtin `plot()` method that will also generate a matplotlib line plot!
